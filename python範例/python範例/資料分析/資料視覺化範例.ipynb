```python
xitun = taichung_traffic_data[taichung_traffic_data['區'] == '西屯區']# 找出西屯區的資料

# 將篩選出來交通事故資料的經緯度放進儲存的變數中
xitun_list = []
for index, row in xitun.iterrows():
  xitun_list.append([row['GPS座標Y'],row['GPS座標X']])

print(xitun_list)
fmap = folium.Map(location=fcu, zoom_start=14)

for d in xitun_list:
  fmap.add_child(folium.Circle(location=d,
              color='red', # Circle 顏色
              radius=10, # Circle 寬度
              fill=True, # 填滿中間區域
              fill_opacity=0.5 # 設定透明度
              ))
fmap
```

## 白天時段

```python
from folium.plugins import HeatMap

xitun = taichung_traffic_data[taichung_traffic_data['區'] == '西屯區']# 找出西屯區的資料

# 將篩選出來交通事故資料的經緯度放進儲存的變數中
xitun_list = []
for index, row in xitun.iterrows():

  xitun_list.append([row['GPS座標Y'],row['GPS座標X']])

fmap = folium.Map(location=fcu, zoom_start=14)
fmap.add_child(HeatMap(data=xitun_list))
fmap
```

```python
from folium.plugins import HeatMap

xitun = taichung_traffic_data[taichung_traffic_data['區'] == '西屯區']# 找出西屯區的資料

# 設定時段 8~20
start = 8
end = 18

# 將篩選出來交通事故資料的經緯度放進儲存的變數中
xitun_list = []
for index, row in xitun.iterrows():
  if row['GPS座標X'] is not None:
    if start <= row["時"] <= end:
      # print(row['月'],row['日'],row['時'], row['分'])
      xitun_list.append([row['GPS座標Y'],row['GPS座標X']])

fmap = folium.Map(location=fcu, zoom_start=14)
fmap.add_child(HeatMap(data=xitun_list))
fmap
```

## 凌晨時段

```python
from folium.plugins import HeatMap

xitun = taichung_traffic_data[taichung_traffic_data['區'] == '西屯區']# 找出西屯區的資料

# 設定時段 23~5
start = 23
end = 5

# 將篩選出來交通事故資料的經緯度放進儲存的變數中
xitun_list = []
for index, row in xitun.iterrows():
  if row['GPS座標X'] is not None:
    if row["時"] >= start or row["時"] <= end: # 小時大於23，或小時小於5
      # print(row["月"], row["日"], row["時"], row["分"], )
      xitun_list.append([row['GPS座標Y'],row['GPS座標X']])

fmap = folium.Map(location=fcu, zoom_start=14)
fmap.add_child(HeatMap(data=xitun_list))
fmap
```
